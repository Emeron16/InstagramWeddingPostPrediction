```python
import requests
import pandas as pd
import time

# Instagram Graph API Configuration
ACCESS_TOKEN = "Access Token"
PAGE_ID = "PageID"  # Replace with your Instagram Business Account ID
API_VERSION = "v21.0"  # Update API version
GRAPH_API_URL = f"https://graph.instagram.com/{API_VERSION}"

# Function to fetch hashtag data
def fetch_instagram_hashtag(hashtag, access_token):
    url = f"{GRAPH_API_URL}//ig_hashtag_search"
    params = {
        "user_id": PAGE_ID,
        "q": hashtag,
        "fields": "id,caption,media_type,media_url,timestamp,like_count,comments_count",
        "access_token": access_token
    }
    response = requests.get(url, params=params)
    print(f"API Request URL: {response.url}")  # Debugging: Print the API request URL
    print(f"Response Status Code: {response.status_code}")
    if response.status_code == 200:
        data = response.json().get("data")
        if data:
            return data
        else:
            print("No hashtag data available.")
            return 
    else:
        print("Error fetching hashtag data:", response.text)
        return 
    
# Function to fetch insights for a media object
def fetch_media_insights(media_id, access_token, metric):
    url = f"{GRAPH_API_URL}/{media_id}/insights"
    params = {
        "metric": metric,  # For example: "impressions,reach"
        "access_token": access_token
    }
    response = requests.get(url, params=params)
    
    if response.status_code == 200:
        insights = response.json().get("data", )
        return {item: item for item in insights}
    else:
        print(f"Error fetching insights for {media_id}:", response.text)
        return {}

# Main function to collect and process Instagram data
def collect_instagram_data():
    hashtag = "weddings"
    print(f"Fetching Instagram posts with #{hashtag} hashtag...")
    hashtag_data = fetch_instagram_hashtag(hashtag, ACCESS_TOKEN)
    all_posts = []

    for post in hashtag_data:
        post_id = post.get("id")
        caption = post.get("caption", "")
        media_type = post.get("media_type", "")
        timestamp = post.get("timestamp", "")
        like_count = post.get("like_count", 0)
        comments_count = post.get("comments_count", 0)

        print(f"Processing post: {post_id}")

        # Fetch insights: impressions and reach
        insights = fetch_media_insights(post_id, ACCESS_TOKEN, "impressions,reach")
        impressions = insights.get("impressions", 0)
        reach = insights.get("reach", 0)

        # Compile post data
        post_info = {
            "Post ID": post_id,
            "Caption": caption,
            "Media Type": media_type,
            "Timestamp": timestamp,
            "Likes": like_count,
            "Comments": comments_count,
            "Impressions": impressions,
            "Reach": reach
        }
        all_posts.append(post_info)
        time.sleep(1)  # To avoid rate limiting

    # Save data to a CSV file
    df = pd.DataFrame(all_posts)
    df.to_csv(f"{hashtag}_posts_data.csv", index=False)
    print(f"Data collection complete. File saved as '{hashtag}_posts_data.csv'.")

# Run the script
if __name__ == "__main__":
    collect_instagram_data()
```
